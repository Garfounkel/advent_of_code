```python
def parse_line(line):
    v = int(line[1:])
    sign = -1 if line[0] == 'L' else 1
    return sign * v

parse_line(lines[0]), parse_line(lines[1]), parse_line(lines[2])
```

```python
turns = [parse_line(line) for line in lines]
turns[:3]
```

```python
move(start, turns[0])
```

```python
def move_all(start, turns):
    count_0 = 0
    
    if start == 0:
        count_0 += 1

    for turn in turns:
        start = move(start, turn)
        if start == 0:
            count_0 += 1

    return count_0

move_all(50, [-68, -30, 48, -5, 60, -55, -1, -99, 14, -82])
```

```python
move_all(50, turns)
```

```python
def move_and_count(start, turn):
    inter = (start + turn)
    full_turns = inter // 100
    res = inter % 100

    full_turns = abs(full_turns)

    if full_turns != 0:
        print(f'{start=}, {turn=}, {inter=}, {full_turns=}, {res=}')

    return abs(full_turns), res

move_and_count(50, 1000)
```

```python
move_and_count(50, 149)
```

```python
def move_all_p2(start, turns):
    count_0 = 0
    
    if start == 0:
        count_0 += 1

    for turn in turns:
        full_turns, start = move_and_count(start, turn)
        count_0 += full_turns

    return count_0


move_all_p2(50, [-68, -30, 48, -5, 60, -55, -1, -99, 14, -82])
```

```python
move_all_p2(50, turns)
```
