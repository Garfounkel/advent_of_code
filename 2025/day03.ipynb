```python
from aocd import get_data

inp = get_data(day=3, year=2025)
inp[:100]
```

```python
example_inp = """
987654321111111
811111111111119
234234234234278
818181911112111
"""
example_inp = example_inp.strip()
example_inp
```

```python
def parse_input(inp):
    return [list(map(int, line)) for line in inp.splitlines()]

parse_input(example_inp)
```

```python
def argmax(line):
    max_value, max_idx = line[0], 0
    for i in range(1, len(line)):
        if line[i] > max_value:
            max_value = line[i]
            max_idx = i
    return max_value, max_idx

argmax(parse_input(example_inp)[0])
```

```python
def max_joltage(line):
    max_value, max_idx = argmax(line)
    if max_idx == len(line) - 1:
        next_max_value, next_max_idx = max_value, max_idx
        max_value, max_idx = argmax(line[:next_max_idx])
    else:
        next_max_value, next_max_idx = argmax(line[max_idx+1:])
    return int(str(max_value) + str(next_max_value))

for line in parse_input(example_inp):
    print(max_joltage(line))
```

```python
def sum_max_joltage(inp):
    lines = parse_input(inp)
    return sum(max_joltage(line) for line in lines)

sum_max_joltage(example_inp)
```

```python
sum_max_joltage(inp)
```

```python
from aocd import submit

submit(sum_max_joltage(inp), part='a', day=3, year=2025)
```
